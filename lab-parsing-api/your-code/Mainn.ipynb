# Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

- Import libraries.

- Define the initial API endpoint URL.

- Make request to obtain data of the 1st page. 
- Flatten the data and store it in an empty object variable.

- Find out the total page count in the 1st page data.

- Use a FOR loop to make requests for the additional pages from 2 to page count. Append the data of each additional page to the flatterned data object.

- Print and review the data you obtained.

```python
response = requests.get(url)   
results = response.json()
results
```

```python
flat_data = json_normalize(results)
flat_data
```

```python

flat_data1 = json_normalize(flat_data.articles[0])

flat_data1
```

```python
# Finding the total page count.
""" This was the first line of the json 
{'total_count': 29002,
 'pagination': {'page_count': 346, 'current_page': 1, 'per_page': 84},"""

total_pages=results['pagination']['page_count']  #accessing by keys  
```

```python
total_pages
```

```python
"""Use a FOR loop to make requests for the additional pages from 2 to page count. 
Append the data of each additional page to the flatterned data object."""

df=pd.DataFrame()
for i in range(2,total_pages):    #from 2 to pagecount
    k=84*i
    url=f'https://www.zalando.fr/api/catalog/articles?categories=promo-homme&limit=84&offset={k}&sort=sale'
    response = requests.get(url)
    results = response.json()
    flattened_data = json_normalize(results)
    flattened_data1 = json_normalize(flattened_data.articles[0])
    flattened_data1=flattened_data1.set_index('sku')
    df = df.append(flattened_data1)

df
```
